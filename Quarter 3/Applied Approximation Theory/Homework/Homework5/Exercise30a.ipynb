```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_coefficient(n: int, k: int):
    return n/k * np.sin(k / n)
    
def series(x: np.ndarray, n: int, N: int):
    sum = np.zeros_like(x)
    sum[:] = 1/np.pi
    for k in range(1, N):
        coeff = get_coefficient(n, k)
        sum +=  2/np.pi * coeff * np.cos(k*x)
    return sum 
```

```python
n = 100
Ns = [x for x in range(1, 5)]
```

```python
fig, ax = plt.subplots(1, 1, figsize = (10, 10))
x = np.linspace(-np.pi, np.pi, 1000)
for i,  N in enumerate( Ns):
    ax.plot(x, series(x, N, n), label = f"N = {N}")
fig.gca().set_facecolor((0.9, 0.9, 0.9))
ax.grid(True)
ax.set_axisbelow(True)
ax.legend()
ax.set_title("Fourier Approximations to $f_N(x)$")
plt.show()
```
