```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def my_integration(f, a, b, n):
    steplength = (b - a) / n
    sum = (f(a) + f(b)) / 2.0

    for i in range(1, n):
        sum += 3*f(i/n)

    return (steplength/3) * sum

def simpson(f, a, b, n):
    steplengh = (b - a)/n
    sum = 0

    for i in range(1, int(n/2)+1):
        first = f(a+(2*i-2) * steplengh)
        second = f(a+ (2*i - 1) * steplengh)
        third = f(a+2*i * steplengh)
        sum += first + 4*second + third
    sum /= 3.
    sum *= steplengh
    return sum

def integrate(f, a, b):
    # the funciton being passed should be the function integrated. 
    sum = f(b) - f(a)
    return sum
```

```python
n = 100
my_error = np.zeros(n)
simp_error = np.zeros(n)


func = lambda x: math.exp(x)

for i in range(1, n+1):
    my_error[i-1] = abs(my_integration(func, 0, 1, i) - integrate(func, 0, 1))
    simp_error[i-1] = abs(simpson(func, 0, 1, i) - integrate(func, 0, 1))

```

```python
plt.plot(my_error, label = "my_error")
plt.plot(simp_error, label = "simp_error")

plt.grid(True)
plt.gca().set_facecolor((0.9, 0.9, 0.9)) 
plt.xlabel("Iteration")
plt.ylabel("Error value")
plt.title(r"Error Difference for $f(x) = e^x$")
plt.legend()
```

```python
n = 100
my_error = np.zeros(n)
simp_error = np.zeros(n)


func = lambda x: math.sqrt(x)
inte = lambda x: 1.5*math.pow(x, 3/2.)

for i in range(1, n+1):
    my_error[i-1] = abs(my_integration(func, 0, 1, i) - integrate(inte, 0, 1))
    simp_error[i-1] = abs(simpson(func, 0, 1, i) - integrate(inte, 0, 1))

```

```python
plt.plot(my_error, label = "my_error")
plt.plot(simp_error, label = "simp_error")

plt.grid(True)
plt.gca().set_facecolor((0.9, 0.9, 0.9)) 
plt.xlabel("Iteration")
plt.ylabel("Error value")
plt.title(r"Error Difference for $f(x) = \sqrt{x}$")
plt.legend()
```
