```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
def psi(x):
    if x < 0 or x > 1:
        return 0
    if x >= 0 and x <= 0.5:
        return 2*x
    return 2*(1 - x)

def xarray(i, k):
    return np.array([2**i * x + k for x in np.linspace(0, 1, 100)])
```

```python
x3 = xarray(0, 0)
x4 = xarray(1, 0)
x5 = xarray(1, 1)
```

```python
xrange = np.linspace(0, 1, 100)

psi1 = np.array([1 for x in xrange])
psi2 = np.array([x for x in xrange])
psi3 = np.array([psi(x) for x in x3])
psi4 = np.array([psi(x) for x in x4])
psi5 = np.array([psi(x) for x in x5])

plt.rcParams['text.usetex'] = True
plt.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'
plt.rcParams["figure.figsize"] = (12, 8)

plt.plot(xrange, psi1, label = r"$\mathcal{X}_1 = 1$")
plt.plot(xrange, psi2, label = r"$\mathcal{X}_2 = x$")
plt.plot(xrange, psi3, label = r"$\mathcal{X}_3 = \psi_{0, 0}$")
plt.plot(xrange, psi4, label = r"$\mathcal{X}_4 = \psi_{1, 0}$")
plt.plot(xrange, psi5, label = r"$\mathcal{X}_5 = \psi_{1, 1}$")
plt.grid(True)
plt.xlabel(r"x")
plt.title(r"First Five Functions of $\mathcal{X}$ on $[0, 1]$")
plt.legend()
plt.show()
```
