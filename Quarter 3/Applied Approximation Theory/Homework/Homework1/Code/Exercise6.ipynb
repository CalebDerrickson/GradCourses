```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rotated(x, theta):
    mat = np.ndarray((2, 2))
    mat[0][0] = np.cos(theta)
    mat[0][1] = -np.sin(theta)
    mat[1][0] = np.sin(theta)
    mat[1][1] = np.cos(theta)

    res = np.matmul(mat, x)
    return res

def expanded_mat(x):
    mat = np.ndarray((2, 2))
    mat[0][0] = 0.5
    mat[0][1] = 0.5
    mat[1][0] = 0.5
    mat[1][1] = 0.5

    res = np.matmul(mat, x)
    return res

def collapse_mat(x):
    mat = np.ndarray((2, 2))
    mat[0][0] = 1
    mat[0][1] = 0
    mat[1][0] = 0
    mat[1][1] = 0

    res = np.matmul(mat, x)
    return res
```

```python
from matplotlib.lines import Line2D 
```

```python
n = 1000
circ = np.zeros((n+1, 2))
expanded = np.zeros((n+1, 2))
collapsed = np.zeros((n+1, 2))
x = np.array([1, 0])
for i in range(1, n+1):
    circ[i, :] = rotated(x, 2*np.pi*i / n)
    expanded[i, :] = expanded_mat(circ[i, :])
    collapsed[i, :] = collapse_mat(circ[i, :])


fig, ax = plt.subplots(1, 1, figsize = (10, 10))

ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])

for point in circ:
    ax.scatter(point[0], point[1], color = 'k', s = 4)
for point in expanded:
    ax.scatter(point[0], point[1], color = 'tab:blue', s = 4)
for point in collapsed:
    ax.scatter(point[0], point[1], color = 'tab:red', s = 4)      
# Custom legend handles and labels
legend_handles = [Line2D([0], [0], marker='o', color='w', markerfacecolor='k', markersize=10),
                  Line2D([0], [0], marker='o', color='w', markerfacecolor='tab:blue', markersize=10),
                  Line2D([0], [0], marker='o', color='w', markerfacecolor='tab:red', markersize=10)]

legend_labels = ['Unit Circle', 'Projection onto (1, 1)', 'Projection onto (1, 0)']

# Adding legend with custom handles and labels
plt.gca().set_facecolor((0.9, 0.9, 0.9)) 
ax.set_axisbelow(True)
ax.legend(legend_handles, legend_labels)
plt.grid(True)
```
