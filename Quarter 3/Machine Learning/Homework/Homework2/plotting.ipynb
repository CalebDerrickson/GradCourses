```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('Problem9/data/accuracies.csv', header=None)
data = data.to_numpy()

plt.rcParams["figure.figsize"] = (12, 8)

plot = plt.subplot(1, 1, 1)
plt.gca().set_facecolor((0.9, 0.9, 0.9)) 

for i, line in enumerate(data):
    if i % 2 == 1:
        plot.plot(100*line, label = f"{round((i+1)/20 * 100, 2)}%")

plt.legend()
plot.grid()
plot.set_axisbelow(True)
plot.set_title("Accuracy of Perceptron")
plot.set_xlabel("Training Round")
plot.set_ylabel("Accuracy (%)")
plt.show()
```
