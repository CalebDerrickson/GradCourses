```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
output_cluster = np.genfromtxt("KmeansData.txt", delimiter=',')
output_cluster
```

```python
colors = ['tab:blue', 'g', 'r']
labels = [r'$C_1$', r'$C_2$', r'$C_3$']

color_mapping = [colors[int(color)] for color in output_cluster[:, 2]]
label_mapping = [labels[int(color)] for color in output_cluster[:, 2]]

plt.rcParams["figure.figsize"] = (12, 8)

plot = plt.subplot(1, 1, 1)
plt.gca().set_facecolor((0.9, 0.9, 0.9)) 

for color, label in zip(colors, labels):
    plot.scatter([], [], c=color, label=label)

plot.grid()
plot.set_axisbelow(True)
plot.scatter(output_cluster[:, 0], output_cluster[:, 1], c = color_mapping, edgecolors='black', linewidths=1)
 
plot.set_title("Gaussian Misture.txt")
plot.set_xlabel("x - value")
plot.set_ylabel("y - value")
plot.legend()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

distortion_pp_10 = pd.read_csv('mnist/outputDistortion_plusplus_10.txt', header=None, sep = ',', names = list(range(201)))
distortion_pp_16 = pd.read_csv('mnist/outputDistortion_plusplus_16.txt', header=None, sep = ',', names = list(range(201)))

distortion_random_10 = pd.read_csv('mnist/outputDistortion_random_10.txt', header=None, sep = ',', names = list(range(201)))
distortion_random_16 = pd.read_csv('mnist/outputDistortion_random_16.txt', header=None, sep = ',', names = list(range(201)))
```

```python
plt.rcParams["figure.figsize"] = (12, 8)

colors = ['tab:red', 'g']
labels = ['KMeans++', 'Random Initialization']

plot = plt.subplot(1, 1, 1)
plt.gca().set_facecolor((0.9, 0.9, 0.9)) 

plt.rcParams["figure.figsize"] = (12, 8)

for color, label in zip(colors, labels):
    #plot.plot([], [], c=color, label=label)
    plot.plot([], [], label=label, color = color)




plot.grid()
plot.set_axisbelow(True)

plotnames = ['KMeans++', 'Random Initialization']
plots = [distortion_pp_10, distortion_random_10]

line_len = np.zeros((2, 20))

for i, dataset in enumerate(plots):
    for j in range(dataset.shape[0]):
        line = dataset.iloc[j].to_numpy()
        line_len[i][j] = np.sum(~np.isnan(line))
        plot.plot(line, color = colors[i])

plot.set_yscale("log")
plot.set_xscale("log")
plot.set_title(r"Distortion difference for KMeans++ and Random Initilization, 10 Clusters.")
plot.set_xlabel("Iteration")
plot.set_ylabel(r"Distorion value $(J_{avg^2})$")
plot.legend()
plt.show()
```

```python
means = [np.mean(arr) for arr in line_len]
stdev = [np.std(arr) for arr in line_len]
means, stdev
```
