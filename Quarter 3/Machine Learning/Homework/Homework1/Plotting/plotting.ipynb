```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
output_cluster = np.genfromtxt("outputCluster.txt", delimiter=',')
output_cluster
```

```python
colors = ['b', 'g', 'r']
labels = [r'$C_1$', r'$C_2$', r'$C_3$']

color_mapping = [colors[int(color)] for color in output_cluster[:, 2]]
label_mapping = [labels[int(color)] for color in output_cluster[:, 2]]

plt.figure(figsize=(8, 6)) 
plt.gca().set_facecolor((0.9, 0.9, 0.9)) 

for color, label in zip(colors, labels):
    plt.scatter([], [], c=color, label=label)

plt.scatter(output_cluster[:, 0], output_cluster[:, 1], c = color_mapping)
 
plt.grid()
plt.title("K means on Toydata.txt")
plt.xlabel("x - value")
plt.ylabel("y - value")
plt.legend()
plt.show()
```
