```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from ipywidgets import interact
%matplotlib inline
```

Problem 1: Exploring pendulum dynamics in the phase plane.

```python
def f(u, t):
    return [u[1], -np.sin(u[0])]

def E(theta, omega):
    return (0.5) * omega**2 - np.cos(theta)
```

```python
@interact(theta0=(-np.pi, np.pi), omega0=(0.0, 1.0), T=(0.1, 25.0))
def plot(theta0, omega0, T):
    
    t = np.arange(0, T, 0.01)
    u0 = (theta0, omega0)
    
    asol = integrate.odeint(f, u0, t)
    
    fig, axs = plt.subplots(3, figsize=(12,8))
    axs[0].plot(t, asol[:,0])
    axs[0].set_ylabel("Theta")
    axs[1].plot(t, asol[:,1])
    axs[1].set_ylabel("Omega")
    
    energy = E(asol[:,0], asol[:,1])
    energy = np.around(energy, 5)
    axs[2].plot(t, energy)
    axs[2].set_ylabel("Energy")
    plt.show()
    return
```

```python
@interact(theta0=(-np.pi, np.pi), omega0=(0.0, 1.0), T=(0.1, 25.0))
def plot(theta0, omega0, T):
    scale = abs(theta0 /np.pi)
    lim = max([np.pi * scale + 0.5, 1])
    t = np.arange(0, T, 0.01)
    u0 = (theta0, omega0)
    fig, ax = plt.subplots(1, figsize=(4,4))
    asol = integrate.odeint(f, u0, t)
    plt.plot(asol[:,0], asol[:,1])
    energy = E(asol[:,0], asol[:,1])
    energy = np.around(energy, 5)
    ax.set_title("Energy: {}".format(energy[0]))
    ax.set_ylim([-lim, lim])
    ax.set_xlim([-lim, lim])
    plt.show()
    return
```

```python
theta_arr = np.arange(0, np.pi - 0.4, 0.2)
omega0 = 0.4
t = np.arange(0, 25, 0.01)
asol_arr = []
energy_arr = []
for theta in theta_arr:
    u0 = [theta, omega0]
    asol = integrate.odeint(f, u0, t)
    asol_arr.append(asol)
    energy = E(asol[:,0], asol[:,1])
    energy = np.around(energy, 4)
    energy_arr.append(energy)
```

```python
delta = 1 / len(theta_arr)
inc = 0
for i in range(len(theta_arr)):
    print(inc)
    inc += delta
```
