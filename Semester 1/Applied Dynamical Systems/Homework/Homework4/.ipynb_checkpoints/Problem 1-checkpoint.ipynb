```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.integrate import solve_ivp
```

```python
def log_eqt(x, t, lm ):
    """
    Returns the Logistic equation for x and lambda
    """
    return lm * x * (1 - x)

def log_eq(x, lm):
    """
    Returns the Logistic equation for x and lambda
    """
    return lm * x * (1 - x)
```

```python
lm = 1 + 2 * math.sqrt(2)

#define time range
timespan = (0., 1.)
init = 1 #number of initial conditions

#initial conditions
x0 = 1

initial_conditions = [x0]

sol = solve_ivp(lambda t, lm, x: log_eqt(x, t, lm), t_span=timespan, y0=initial_conditions, max_step=0.01, args=(lm, ))
```

```python
y = sol.y[0]
plt.plot(y)
```

```python
timespan = (0., 1.)

for x0 in np.linspace(-2, 2, 50):
    initial_conditions = [x0]
    sol = solve_ivp(lambda t, lm, x: log_eqt(x, t, lm), t_span=timespan, y0=initial_conditions, max_step=0.01, args=(lm,))

    plt.plot(sol.y[0][:60])
```
