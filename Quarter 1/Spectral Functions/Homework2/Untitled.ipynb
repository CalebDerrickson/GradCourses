```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create a path graph
G = nx.path_graph(20)

# Getting Laplacian matrix
L = nx.laplacian_matrix(G).toarray()

# Getting the second eigenvector
v_2 = np.linalg.eigh(L)[1][:, 1]

# Sweep cut with 2 clusters
num_clusters = 2
cluster_assignments = np.zeros(len(G.nodes()))

# Find the threshold that separates the clusters
threshold = np.median(v_2)

# Assign nodes to clusters based on the second eigenvector
cluster_assignments[v_2 <= threshold] = 1
cluster_assignments[v_2 > threshold] = 2

# Plot the graph with nodes colored by cluster
pos = nx.spring_layout(G)
nx.draw(G, pos, node_color=cluster_assignments, cmap=plt.cm.Blues, with_labels=True)
plt.show()
```
